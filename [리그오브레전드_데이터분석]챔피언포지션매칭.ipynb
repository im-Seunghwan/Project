```
def position_win_table(match_id):
  match = f"https://kr.api.riotgames.com/lol/match/v4/matches/{match_id}" +'?api_key=' + key
  timeline =  f"https://kr.api.riotgames.com/lol/match/v4/timelines/by-match/{match_id}" +'?api_key=' + key

  r = requests.get(match)
  
  if (r.status_code == 200):
    m = r.json()
  elif (r.status_code == 429):
    time.sleep(int(r.headers["Retry-After"]))
    r = requests.get(match)
    m = r.json()
  elif r.status_code == 403:
    print('need new API')
    raise('Error')
  
  r2 = requests.get(timeline)

  if r2.status_code == 200:
    t = r2.json()
  elif r2.status_code == 429:
    time.sleep(int(r2.headers["Retry-After"]))
    r2 = requests.get(timeline)
    t= r2.json()
  elif r2.status_code == 403:
    print('need new API')
    raise('Error')
  
  # 각 플레이어 넘버에 해당하는 포지션이 dict 형태로 나옴 , key는 1~10
  
  position_dict = roleml.predict(m, t)

  try:
    if len(position_dict) == 10:

      for i in range(1,11):
        position = position_dict[i]
      
        if i <= 5:
          if position == 'top':
            t1_top_champion_code = m['participants'][i-1]['championId'] #챔피언 코드로
            t1_top_name = chp_match_dict[t1_top_champion_code] # chp_match_dict = {1:'챔피언이름'}
            t1_top_tier = tier['top'][t1_top_name]
          elif position == 'jungle':
            t1_jug_champion_code = m['participants'][i-1]['championId']
            t1_jug_name = chp_match_dict[t1_jug_champion_code]
            t1_jug_tier = tier['jug'][t1_jug_name]
          elif position == 'mid':
            t1_mid_champion_code = m['participants'][i-1]['championId']
            t1_mid_name = chp_match_dict[t1_mid_champion_code]
            t1_mid_tier = tier['mid'][t1_mid_name]
          elif position == 'bot':
            t1_adc_champion_code = m['participants'][i-1]['championId']
            t1_adc_name = chp_match_dict[t1_adc_champion_code]
            t1_adc_tier = tier['adc'][t1_adc_name] 
          elif position == 'supp':
            t1_sup_champion_code = m['participants'][i-1]['championId']
            t1_sup_name = chp_match_dict[t1_sup_champion_code] 
            t1_sup_tier = tier['sup'][t1_sup_name]
        else:
          if position == 'top':
            t2_top_champion_code = m['participants'][i-1]['championId']
            t2_top_name = chp_match_dict[t2_top_champion_code]
            t2_top_tier = tier['top'][t2_top_name]
          elif position == 'jungle':
            t2_jug_champion_code = m['participants'][i-1]['championId']
            t2_jug_name = chp_match_dict[t2_jug_champion_code]
            t2_jug_tier = tier['jug'][t2_jug_name]
          elif position == 'mid':
            t2_mid_champion_code = m['participants'][i-1]['championId']
            t2_mid_name = chp_match_dict[t2_mid_champion_code]
            t2_mid_tier = tier['mid'][t2_mid_name] 
          elif position == 'bot':
            t2_adc_champion_code = m['participants'][i-1]['championId']
            t2_adc_name = chp_match_dict[t2_adc_champion_code]
            t2_adc_tier = tier['adc'][t2_adc_name]
          elif position == 'supp':
            t2_sup_champion_code = m['participants'][i-1]['championId']
            t2_sup_name = chp_match_dict[t2_sup_champion_code]
            t2_sup_tier = tier['sup'][t2_sup_name] 

    t1_top_champion = chp_match_dict[t1_top_champion_code] # chp_match_dict = {1:'garen'}
    t1_jug_champion = chp_match_dict[t1_jug_champion_code]
    t1_mid_champion = chp_match_dict[t1_mid_champion_code]
    t1_adc_champion = chp_match_dict[t1_adc_champion_code]
    t1_sup_champion = chp_match_dict[t1_sup_champion_code]

    t2_top_champion = chp_match_dict[t2_top_champion_code]
    t2_jug_champion = chp_match_dict[t2_jug_champion_code]
    t2_mid_champion = chp_match_dict[t2_mid_champion_code]
    t2_adc_champion = chp_match_dict[t2_adc_champion_code]
    t2_sup_champion = chp_match_dict[t2_sup_champion_code]

    top_counter_info = counter['top'][f'{t1_top_champion}-{t2_top_champion}'] 
    jug_counter_info = counter['jug'][f'{t1_jug_champion}-{t2_jug_champion}']
    mid_counter_info = counter['mid'][f'{t1_mid_champion}-{t2_mid_champion}']
    adc_counter_info = counter['adc'][f'{t1_adc_champion}-{t2_adc_champion}']
    sup_counter_info = counter['sup'][f'{t1_sup_champion}-{t2_sup_champion}']

    result = m['teams'][0]['win']
    new_data = {'MATCH_ID':match_id, 'T1_top': t1_top_tier , 'T1_jug': t1_jug_tier , 'T1_mid': t1_mid_tier, 'T1_adc': t1_adc_tier, 'T1_sup': t1_sup_tier,
                'T2_top': t2_top_tier, 'T2_jug': t2_jug_tier, 'T2_mid': t2_mid_tier, 'T2_adc': t2_adc_tier, 'T2_sup': t2_sup_tier , 
                'top_counter': top_counter_info, 'jug_counter': jug_counter_info, 'mid_counter': mid_counter_info, 'adc_counter':adc_counter_info, 'sup_counter':sup_counter_info,
                'win':result}
    return new_data
  except:
    print('error')
```
